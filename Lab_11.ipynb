# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python

def tsp(graph):
    n = len(graph)
    dp = [[None]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0], [0]

        if dp[city][mask] is not None:
            return dp[city][mask]

        min_cost = float('inf')
        best_path = []

        for next_city in range(n):
            if not (mask & (1 << next_city)):
                cost, tail_path = visit(next_city, mask | (1 << next_city))
                total_cost = graph[city][next_city] + cost

                if total_cost < min_cost:
                    min_cost = total_cost
                    best_path = [next_city] + tail_path

        dp[city][mask] = (min_cost, best_path)
        return min_cost, best_path

    cost, path_suffix = visit(0, 1)
    full_path = [0] + path_suffix
    return cost, full_path
```

```python

graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

cost, path = tsp(graph)
print("Minimum travelling cost:", cost)
print("Optimal Path:", path)
```
