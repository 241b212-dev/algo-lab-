# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

````python

def tsp(graph):
    n = len(graph)
    dp = [[None]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0], [0]

        if dp[city][mask] is not None:
            return dp[city][mask]

        min_cost = float('inf')
        best_path = []

        for next_city in range(n):
            if not (mask & (1 << next_city)):
                cost, tail_path = visit(next_city, mask | (1 << next_city))
                total_cost = graph[city][next_city] + cost

                if total_cost < min_cost:
                    min_cost = total_cost
                    best_path = [next_city] + tail_path

        dp[city][mask] = (min_cost, best_path)
        return min_cost, best_path

    cost, path_suffix = visit(0, 1)
    full_path = [0] + path_suffix
    return cost, full_path
````

````python

graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

cost, path = tsp(graph)
print("Minimum travelling cost:", cost)
print("Optimal Path:", path)
````

````python
# Travelling Salesman Problem (TSP) using Dynamic Programming – Exercise Solutions

## 1. Print Path of Cities Visited

```python
# TSP with path printing

def tsp_with_path(graph):
    n = len(graph)
    dp = [[-1] * (1 << n) for _ in range(n)]
    path = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        # If all cities are visited, return cost to go back to source city
        if mask == (1 << n) - 1:
            return graph[city][0]

        # If already calculated
        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        next_node = -1

        for next_city in range(n):
            # Check if city is not visited
            if mask & (1 << next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))

                if new_cost < ans:
                    ans = new_cost
                    next_node = next_city

        dp[city][mask] = ans
        path[city][mask] = next_node
        return ans

    minimum_cost = visit(0, 1)

    # Reconstruct path
    city = 0
    mask = 1
    tour = [0]

    while True:
        next_city = path[city][mask]

        if next_city == -1:
            break

        tour.append(next_city)
        mask |= (1 << next_city)
        city = next_city

    tour.append(0)  # Return to starting city

    return minimum_cost, tour


# Example graph
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

cost, tour = tsp_with_path(graph)

print("Minimum travelling cost:", cost)
print("Path followed:", tour)
```

### Output

```python
Minimum travelling cost: 80
Path followed: [0, 1, 3, 2, 0]
```

---

# 2. Take Graph Input from User

```python
# TSP with user input

def tsp(graph):
    n = len(graph)
    dp = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1 << n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1 << next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))
                ans = min(ans, new_cost)

        dp[city][mask] = ans
        return ans

    return visit(0, 1)


# Input from user
n = int(input("Enter number of cities: "))

graph = []

print("Enter adjacency matrix:")
for i in range(n):
    row = list(map(int, input().split()))
    graph.append(row)

print("Minimum travelling cost:", tsp(graph))
```

### Sample Input

```python
Enter number of cities: 4
Enter adjacency matrix:
0 10 15 20
10 0 35 25
15 35 0 30
20 25 30 0
```

### Sample Output

```python
Minimum travelling cost: 80
```

---

# 3. Test with Different Number of Cities

## Example with 3 Cities

```python
graph = [
    [0, 20, 42],
    [20, 0, 30],
    [42, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

### Output

```python
Minimum travelling cost: 92
```

---

## Example with 5 Cities

```python
graph = [
    [0, 14, 4, 10, 20],
    [14, 0, 7, 8, 7],
    [4, 5, 0, 7, 16],
    [11, 7, 9, 0, 2],
    [18, 7, 17, 4, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

### Output

```python
Minimum travelling cost: 30
```

---

# Conclusion

* Dynamic Programming reduces repeated calculations in TSP.
* Bit Masking is used to track visited cities efficiently.
* The algorithm finds the minimum travelling cost and optimal path.
* Time Complexity: O(n² × 2ⁿ)
````
