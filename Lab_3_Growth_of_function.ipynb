# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1
import matplotlib.pyplot as plt

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1) # The time taken is constant, regardless of 'n'

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
import matplotlib.pyplot as plt
import math

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0
    while low <= high:
        steps += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return steps  # Return steps to simulate time
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps # Target not found, but we still count steps

n_values = [2**i for i in range(1, 15)] # Input sizes that make log n clear
time_values = []

for n in n_values:
    data = list(range(n))
    # For simplicity, we'll search for the last element or assume target is found
    # The number of steps will be proportional to log(n)
    # A more rigorous simulation would run binary_search and record actual steps
    # For demonstrating O(log n) we can directly use math.log2(n)
    if n > 0:
        time_values.append(math.log2(n))
    else:
        time_values.append(0)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
import matplotlib.pyplot as plt
import math

def merge_sort_operations(n):
    # Simulate operations for merge sort, which is O(n log n)
    # For simplicity, we'll return n * log2(n) as the 'time'
    # assuming the base of the logarithm is 2 for binary splitting.
    if n <= 1:
        return 0
    return n * math.log2(n)

n_values = list(range(1, 1001, 50)) # Input sizes from 1 to 1000, step 50
time_values = []

for n in n_values:
    time_values.append(merge_sort_operations(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='O(n log n) - Merge Sort')
plt.xlabel("Input size n")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity")
plt.grid(True)
plt.legend()
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6: Take a suitable example and write a function that runs in Polynomial Time Complexity (e.g., O(n^2)). Plot the graph.

import matplotlib.pyplot as plt

def polynomial_time_example(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    time_values.append(polynomial_time_example(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='O(n^2) - Polynomial Time Complexity')
plt.xlabel("Input size n")
plt.ylabel("Operations (n^2)")
plt.title("O(n^2) Polynomial Time Complexity")
plt.grid(True)
plt.legend()
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7: Take a suitable example and write a function that runs in Exponential Time Complexity (e.g., O(c^n)). Plot the graph as well, similar to above example.

import matplotlib.pyplot as plt

def exponential_fibonacci(n):
    if n <= 1:
        return n
    return exponential_fibonacci(n - 1) + exponential_fibonacci(n - 2)

def fib_operations(n):
    # This function simulates the number of operations for exponential_fibonacci(n)
    # It's an approximation, as the actual number of calls grows exponentially.
    # For small n, it's roughly proportional to (1.618)^n
    if n <= 1:
        return 1
    # Using math.pow for a clear exponential representation for plotting
    # The base is the golden ratio approx 1.618, often associated with Fibonacci growth
    return (1.6180339887)**n

n_values = list(range(1, 21)) # Exponential growth is very fast, so keep n small
time_values = []

for n in n_values:
    time_values.append(fib_operations(n))

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, label='O(c^n) - Exponential Time Complexity (Fibonacci)')
plt.xlabel("Input size n")
plt.ylabel("Operations (c^n)")
plt.title("O(c^n) Exponential Time Complexity")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

# Define functions for each time complexity
def O_1(n): # Constant Time
    return 1

def O_log_n(n): # Logarithmic Time
    return math.log2(n) if n > 0 else 0

def O_n(n): # Linear Time
    return n

def O_n_log_n(n): # Log-Linear Time
    return n * math.log2(n) if n > 0 else 0

def O_n_squared(n): # Polynomial Time (Quadratic)
    return n**2

def O_2_power_n(n): # Exponential Time
    return 2**n

# Input sizes for plotting (keeping it small for exponential complexity)
n_values = list(range(1, 25))

# Calculate time values for each complexity
time_O_1 = [O_1(n) for n in n_values]
time_O_log_n = [O_log_n(n) for n in n_values]
time_O_n = [O_n(n) for n in n_values]
time_O_n_log_n = [O_n_log_n(n) for n in n_values]
time_O_n_squared = [O_n_squared(n) for n in n_values]
time_O_2_power_n = [O_2_power_n(n) for n in n_values]

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(n_values, time_O_1, label='O(1) - Constant', linestyle='--')
plt.plot(n_values, time_O_log_n, label='O(log n) - Logarithmic')
plt.plot(n_values, time_O_n, label='O(n) - Linear')
plt.plot(n_values, time_O_n_log_n, label='O(n log n) - Log-Linear')
plt.plot(n_values, time_O_n_squared, label='O(n^2) - Quadratic')
plt.plot(n_values, time_O_2_power_n, label='O(2^n) - Exponential', color='red', linestyle='-.')

plt.xlabel("Input size n")
plt.ylabel("Operations (Time)")
plt.title("Comparison of Time Complexities")
plt.grid(True)
plt.legend()
plt.ylim(0, 1000) # Limit y-axis to better visualize lower complexities
plt.show()
```
